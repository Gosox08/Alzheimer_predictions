```python
import pandas as pd
import numpy as np
import os
import cv2
import pickle
import matplotlib.pyplot as plt
import random
```

```python
X = pd.DataFrame
```

```python
datadir_train = "train"
datadir_test = "test"
categories = os.listdir(path='/content/drive/MyDrive/Alzheimer_s Dataset/test')
```

```python
def creating_data(datadir, saved_data):
  for category in categories:
    path = os.path.join('/content/drive/MyDrive/Alzheimer_s Dataset', datadir, category)
    for img in os.listdir(path):
      try:
        img_array = cv2.imread(os.path.join(path, img),cv2.IMREAD_GRAYSCALE)
        saved_data.append([img_array, category])
        random.shuffle(saved_data)
      except Exception as e:
        print("Photo probably is broken.")
```

```python
def creating_set(X, y):
  for features, labels in training_data:
    X.append(features)
    y.append(labels)
  X = np.array(X)
  y = np.array(y)
```

```python
training_data=[]
creating_data(datadir_train, training_data)
X_train=[]
y_train=[]
creating_set(X_train, y_train)
```

```python
testing_data=[]
creating_data(datadir_test, testing_data)
X_test=[]
y_test=[]
creating_set(X_test, y_test)
```

```python
plt.imshow(X_train[0])
print(y_train[0])
```

```python
def saving_data(X, y, X_file, y_file):
  pickle_out = open(X_file, "wb")
  pickle.dump(X, pickle_out)
  pickle_out.close()

  pickle_out = open(y_file, "wb")
  pickle.dump(y, pickle_out)
  pickle_out.close()
```

```python
saving_data(X_train, y_train, "X_train.pickle",  "y_train.pickle")
```

```python
saving_data(X_test, y_test, "X_test.pickle", "y_test.pickle")
```
